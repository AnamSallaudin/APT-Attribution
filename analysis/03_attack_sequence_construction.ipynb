```python
import pandas as pd

labels = pd.read_csv("../data/processed/labels.csv")

labels.head()
```

```python
labels['APT-group'].value_counts()
```

```python
apt_attack_map = {
    "APT28": ["T1059", "T1027", "T1105", "T1071", "T1055", "T1082"],
    "APT29": ["T1566", "T1204", "T1059", "T1027", "T1071"],
    "APT1": ["T1059", "T1047", "T1086", "T1105"],
    "WINNTI": ["T1059", "T1027", "T1055", "T1071", "T1105"],
    "EQUATIONGROUP": ["T1055", "T1068", "T1027", "T1105"],
    "APT10": ["T1059", "T1086", "T1027", "T1071"],
    "DARKHOTEL": ["T1566", "T1204", "T1059", "T1105"],
    "GORGONGROUP": ["T1059", "T1027", "T1071", "T1105"]
}
```

```python
import random

def generate_sequence(techniques, min_len=3, max_len=6):
    length = random.randint(min_len, min(max_len, len(techniques)))
    sequence = random.sample(techniques, length)
    return sequence
```

```python
random.seed(42)

labels["attack_sequence"] = labels["APT-group"].apply(
    lambda apt: generate_sequence(apt_attack_map[apt])
)

labels.head()
```

```python
labels.to_csv(
    "../attack_sequences/apt_attack_sequences.csv",
    index=False
)
```
