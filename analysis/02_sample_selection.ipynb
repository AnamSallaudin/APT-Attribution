```python
import pandas as pd

labels = pd.read_csv("../data/processed/labels.csv")

# sanity check
labels['APT-group'].value_counts()
```

```python
SAMPLES_PER_GROUP = 50

selected = (
    labels
    .groupby("APT-group", group_keys=False)
    .apply(lambda x: x.sample(
        n=min(len(x), SAMPLES_PER_GROUP),
        random_state=42
    ))
)

selected.shape
```

```python
selected.to_csv(
    "../data/processed/selected_samples.csv",
    index=False
)
```
