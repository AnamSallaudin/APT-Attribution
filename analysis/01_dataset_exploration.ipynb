```python
df = df.dropna(subset=['APT-group', 'SHA256'])
```

```python
df['APT-group'] = (
    df['APT-group']
    .str.upper()
    .str.replace("-", "", regex=False)
    .str.replace(" ", "", regex=False)
)
```

```python
selected_apts = [
    'GORGONGROUP',
    'APT1',
    'WINNTI',
    'EQUATIONGROUP',
    'APT10',
    'DARKHOTEL',
    'APT29',
    'APT28'
]

df = df[df['APT-group'].isin(selected_apts)]
```

```python
df['APT-group'].value_counts()
```

```python
df[['ID', 'SHA256', 'APT-group', 'Country']].to_csv(
    "../data/processed/labels.csv", index=False
)
```
